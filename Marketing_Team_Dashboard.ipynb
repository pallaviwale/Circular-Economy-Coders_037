```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


sns.set(style="whitegrid")
```

```python
customers_df = pd.read_csv('olist_customers_dataset.csv')
orders_df = pd.read_csv('olist_orders_dataset.csv')
order_items_df = pd.read_csv('olist_order_items_dataset.csv')
order_reviews_df = pd.read_csv('olist_order_reviews_dataset.csv')
sellers_df = pd.read_csv('olist_sellers_dataset.csv')
products_df = pd.read_csv('olist_products_dataset.csv')
```

# Customer distribution by state

```python
customer_distribution_state = customers_df.groupby('customer_state').size().reset_index(name='customer_count')


plt.figure(figsize=(12, 6))
sns.barplot(data=customer_distribution_state, x='customer_state', y='customer_count', palette='viridis')
plt.title('Customer Distribution by State')
plt.xticks(rotation=45)
plt.show()
```

#  Average review score per product


```python
review_product_df = pd.merge(order_reviews_df, order_items_df, on='order_id')

average_review_score_per_product = review_product_df.groupby('product_id')['review_score'].mean().reset_index()

plt.figure(figsize=(12, 6))
sns.histplot(average_review_score_per_product['review_score'], bins=20, kde=True, color='skyblue')
plt.title('Average Review Score per Product')
plt.xlabel('Review Score')
plt.ylabel('Frequency')
plt.show()
```

# Average freight value by state


```python
orders_customers_df = pd.merge(orders_df, customers_df, on='customer_id')
orders_items_customers_df = pd.merge(orders_customers_df, order_items_df, on='order_id')


average_freight_value_by_state = orders_items_customers_df.groupby('customer_state')['freight_value'].mean().reset_index()


plt.figure(figsize=(12, 6))
sns.barplot(data=average_freight_value_by_state, x='customer_state', y='freight_value', palette='coolwarm')
plt.title('Average Freight Value by State')
plt.xticks(rotation=45)
plt.ylabel('Average Freight Value')
plt.show()
```

# Average review score by seller


```python
average_review_score_by_seller = review_product_df.groupby('seller_id')['review_score'].mean().reset_index()

plt.figure(figsize=(12, 6))
sns.histplot(average_review_score_by_seller['review_score'], bins=20, kde=True, color='lightgreen')
plt.title('Average Review Score by Seller')
plt.xlabel('Review Score')
plt.ylabel('Frequency')
plt.show()
```

# The number of orders by state


```python
orders_by_state = orders_customers_df.groupby('customer_state').size().reset_index(name='order_count')


plt.figure(figsize=(12, 6))
sns.barplot(data=orders_by_state, x='customer_state', y='order_count', palette='magma')
plt.title('Number of Orders by State')
plt.xticks(rotation=45)
plt.ylabel('Number of Orders')
plt.show()
```

# Distribution of freight values


```python
plt.figure(figsize=(10, 5))
sns.histplot(orders_items_customers_df['freight_value'], bins=30, color='coral', kde=True)
plt.title('Distribution of Freight Values')
plt.xlabel('Freight Value')
plt.ylabel('Frequency')
plt.show()
```

# Top 10 products by the number of orders


```python
top_products = order_items_df['product_id'].value_counts().head(10)

plt.figure(figsize=(10, 5))
sns.barplot(x=top_products.index, y=top_products.values, palette='plasma')
plt.title('Top 10 Products by Number of Orders')
plt.xlabel('Product ID')
plt.ylabel('Number of Orders')
plt.show()
```

# Top 10 sellers by number of orders


```python
top_sellers = order_items_df['seller_id'].value_counts().head(10)

plt.figure(figsize=(10, 6))
sns.barplot(x=top_sellers.index, y=top_sellers.values, palette='muted')
plt.title('Top 10 Sellers by Number of Orders')
plt.xlabel('Seller ID')
plt.ylabel('Number of Orders')
plt.xticks(rotation=45)
plt.show()

# OPTional
top_sellers_df = pd.DataFrame({'seller_id': top_sellers.index, 'number_of_orders': top_sellers.values})
print("Top 10 Sellers by Number of Orders:\n", top_sellers_df)
```

# Distribution of product prices


```python
plt.figure(figsize=(10, 6))
sns.histplot(order_items_df['price'], bins=30, color='blue', kde=True)
plt.title('Distribution of Product Prices')
plt.xlabel('Product Price')
plt.ylabel('Frequency')
plt.show()
```

# Selecting relevant numeric columns for correlation analysis


```python
corr_df = order_items_df[['price', 'freight_value']].join(order_reviews_df[['review_score']])

correlation_matrix = corr_df.corr()

plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap of Numeric Features')
plt.show()
```

# Average delivery time by state


```python
orders_df['order_purchase_timestamp'] = pd.to_datetime(orders_df['order_purchase_timestamp'])
orders_df['order_delivered_customer_date'] = pd.to_datetime(orders_df['order_delivered_customer_date'])


orders_df['delivery_time'] = (orders_df['order_delivered_customer_date'] - orders_df['order_purchase_timestamp']).dt.days


orders_customers_delivery_df = pd.merge(orders_df, customers_df, on='customer_id')

avg_delivery_time_by_state = orders_customers_delivery_df.groupby('customer_state')['delivery_time'].mean().reset_index()

plt.figure(figsize=(12, 6))
sns.barplot(data=avg_delivery_time_by_state, x='customer_state', y='delivery_time', palette='Blues_d')
plt.title('Average Delivery Time by State (in Days)')
plt.xticks(rotation=45)
plt.ylabel('Average Delivery Time (Days)')
plt.show()
```

# Distribution of average freight value per seller


```python
avg_freight_by_seller = order_items_df.groupby('seller_id')['freight_value'].mean().reset_index()

plt.figure(figsize=(10, 6))
sns.histplot(avg_freight_by_seller['freight_value'], bins=30, kde=True, color='purple')
plt.title('Distribution of Average Freight Value by Seller')
plt.xlabel('Freight Value')
plt.ylabel('Frequency')
plt.show()

print("Freight Value Statistics by Seller:\n", avg_freight_by_seller['freight_value'].describe())
```

# Monthly Order Volume Over Time


```python
orders_df['order_purchase_month'] = orders_df['order_purchase_timestamp'].dt.to_period('M').astype(str)


orders_per_month = orders_df.groupby('order_purchase_month').size().reset_index(name='order_count')

plt.figure(figsize=(12, 6))
sns.lineplot(x='order_purchase_month', y='order_count', data=orders_per_month, marker='o', color='coral')
plt.title('Monthly Order Volume Over Time')
plt.xticks(rotation=45)
plt.ylabel('Number of Orders')
plt.xlabel('Month')
plt.show()
```

## Top 5 Categories with Highest Average Price

```python
products_items_df = pd.merge(order_items_df, products_df, on='product_id')

avg_price_per_category = products_items_df.groupby('product_category_name')['price'].mean().reset_index()


top_categories_by_price = avg_price_per_category.sort_values(by='price', ascending=False).head(5)

plt.figure(figsize=(10, 6))
sns.barplot(x='product_category_name', y='price', data=top_categories_by_price, palette='Set2')
plt.title('Top 5 Product Categories with Highest Average Price')
plt.ylabel('Average Price')
plt.xlabel('Product Category')
plt.xticks(rotation=45)
plt.show()

print("Top 5 Categories by Average Price:\n", top_categories_by_price)
```

## Review Score Distribution by Month

```python
orders_reviews_df['order_purchase_timestamp'] = pd.to_datetime(orders_reviews_df['order_purchase_timestamp'])
orders_reviews_df['review_score'] = orders_reviews_df['review_score'].astype(int)

orders_reviews_df['purchase_month'] = orders_reviews_df['order_purchase_timestamp'].dt.to_period('M').astype(str)

avg_review_score_by_month = orders_reviews_df.groupby('purchase_month')['review_score'].mean().reset_index()

plt.figure(figsize=(12, 6))
sns.lineplot(x='purchase_month', y='review_score', data=avg_review_score_by_month, marker='o', color='seagreen')
plt.title('Average Review Score by Month')
plt.xticks(rotation=45)
plt.ylabel('Average Review Score')
plt.xlabel('Month')
plt.show()
```
